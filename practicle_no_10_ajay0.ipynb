To study and understand Version Control using Git and GitHub, and perform basic
operations like adding, committing, and pushing files to a repository.
Theory
Version Control System (VCS) is used to track changes in files and manage different
versions of a project. It helps multiple developers work together efficiently.
Git is a distributed version control system that tracks changes locally on your system.
GitHub is a cloud-based platform that hosts Git repositories and allows collaboration.
Basic workflow in Git:
• Create repository
• Add files
• Stage files (git add)
• Commit changes (git commit)
• Push to remote repository (git push)
